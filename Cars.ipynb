```python
# create function to search a column and replace ? values with mean value
def replace_missing_values(column):
    total = 0
    counter = 0
    for i in range(len(cars_df)):
        if cars_df[column][i] != '?':
            counter += 1
            total += float(cars_df[column][i])
        else:
            continue
    avg_value = round(total/counter,0)

    # assign the average value to records where price == ?
    for i in range(len(cars_df)):
        if cars_df[column][i] == '?':
            cars_df[column][i] = avg_value
        else:
            continue
    return(cars_df[column].value_counts())
```

```python
# replace instances where price == ? with mean price
replace_missing_values('price')
```

```python
# replace missing values for bore,stroke, horsepower, peak-rpm
replace_missing_values('bore')
replace_missing_values('stroke')
replace_missing_values('horsepower')
```

```python
# update price, bore, stroke, horsepower, peak-rpm data types to float or int
cars_df.price = cars_df.price.astype(np.int64)
cars_df.bore = cars_df.bore.astype(np.float64)
cars_df.stroke = cars_df.stroke.astype(np.float64)
cars_df.horsepower = cars_df.horsepower.astype(np.int64)
cars_df.dtypes
```

```python
# import libraries needed for regression, classification tree, & neural net
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# create a multiple linear regression model
# find the relationship between city-mpg, highway-mpg, horsepower, and price
# return the intercept and coefficient
X = cars_df[['horsepower','city-mpg','highway-mpg']]
Y = cars_df[['price']]
mlr = LinearRegression()
mlr.fit(X,Y)
mlr_intercept = mlr.intercept_
mlr_coef = mlr.coef_
print('The y-intercept of the regression equation is: ', mlr_intercept, sep='')
print('The coefficients of the regression equation are: ', mlr_coef, sep='')
```
